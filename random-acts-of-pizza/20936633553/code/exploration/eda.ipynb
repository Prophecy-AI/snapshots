```
import json
import pandas as pd
import numpy as np

# Load the training data
with open('/home/data/train.json', 'r') as f:
    train_data = json.load(f)

print(f"Number of training samples: {len(train_data)}")
print(f"Type of data: {type(train_data)}")
print(f"First sample keys: {list(train_data[0].keys())}")
```

```
# Convert to DataFrame for easier analysis
df = pd.DataFrame(train_data)

# Check target distribution
print("Target distribution:")
print(df['requester_received_pizza'].value_counts())
print(f"\nPercentage of positive class: {df['requester_received_pizza'].mean():.3f}")

# Check data types
print(f"\nDataFrame shape: {df.shape}")
print(f"\nData types:")
print(df.dtypes.value_counts())
```

```
# Examine text features
print("Text feature analysis:")
print(f"Request title length (chars): {df['request_title'].str.len().describe()}")
print(f"\nRequest text length (chars): {df['request_text'].str.len().describe()}")
print(f"\nRequest text length (words): {df['request_text'].str.split().str.len().describe()}")

# Check some key metadata features
print("\n\nKey metadata features:")
print(f"Account age at request (days): {df['requester_account_age_in_days_at_request'].describe()}")
print(f"\nNumber of comments at request: {df['requester_number_of_comments_at_request'].describe()}")
print(f"\nUpvotes minus downvotes at request: {df['requester_upvotes_minus_downvotes_at_request'].describe()}")

# Check user flair distribution
print(f"\nUser flair distribution:")
print(df['requester_user_flair'].value_counts())
```
