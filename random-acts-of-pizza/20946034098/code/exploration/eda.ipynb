```
import pandas as pd
import json
import numpy as np

# Load the training data
train_path = "/home/data/train.json"

# Read the JSON file
with open(train_path, 'r') as f:
    train_data = json.load(f)

print(f"Number of training samples: {len(train_data)}")
print(f"Type of data: {type(train_data)}")
print(f"First sample keys: {list(train_data[0].keys())}")
```

```
# Convert to DataFrame for easier analysis
df = pd.DataFrame(train_data)
print(f"DataFrame shape: {df.shape}")
print("\nColumn names:")
print(df.columns.tolist())
```

```
# Analyze target distribution
target = df['requester_received_pizza']
print("Target distribution:")
print(target.value_counts())
print(f"\nClass balance:")
print(target.value_counts(normalize=True))

# Check for missing values
print(f"\nMissing values per column:")
print(df.isnull().sum().sort_values(ascending=False).head(10))
```

```
# Analyze text features
print("Text feature analysis:")
print(f"\nRequest title length stats:")
print(df['request_title'].apply(len).describe())

print(f"\nRequest text length stats:")
print(df['request_text'].apply(len).describe())

print(f"\nRequest text edit aware length stats:")
print(df['request_text_edit_aware'].apply(len).describe())

# Check for unique values in categorical features
print(f"\nUnique user flair values:")
print(df['requester_user_flair'].value_counts())

print(f"\nPost edited distribution:")
print(df['post_was_edited'].value_counts())
```

```
# Analyze numerical features
numerical_cols = [
    'number_of_downvotes_of_request_at_retrieval',
    'number_of_upvotes_of_request_at_retrieval', 
    'request_number_of_comments_at_retrieval',
    'requester_account_age_in_days_at_request',
    'requester_number_of_comments_at_request',
    'requester_number_of_posts_at_request',
    'requester_number_of_subreddits_at_request',
    'requester_upvotes_minus_downvotes_at_request',
    'requester_upvotes_plus_downvotes_at_request'
]

print("Numerical feature statistics:")
for col in numerical_cols:
    print(f"\n{col}:")
    print(df[col].describe())
```
