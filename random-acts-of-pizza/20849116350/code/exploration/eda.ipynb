```
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import seaborn as sns

# Load the training data
with open('/home/data/train.json', 'r') as f:
    train_data = json.load(f)

print(f"Number of training samples: {len(train_data)}")
print(f"Type of data: {type(train_data)}")
print(f"First sample keys: {list(train_data[0].keys())}")
```

```
# Convert to DataFrame for easier analysis
df = pd.DataFrame(train_data)

# Target distribution
print("Target distribution:")
print(df['requester_received_pizza'].value_counts())
print(f"Success rate: {df['requester_received_pizza'].mean():.3f}")

# Basic statistics on key features
print("\nKey numerical features:")
key_features = ['requester_account_age_in_days_at_request', 
                'requester_number_of_comments_at_request',
                'requester_number_of_posts_at_request',
                'requester_upvotes_minus_downvotes_at_request']

for feature in key_features:
    print(f"{feature}: mean={df[feature].mean():.1f}, median={df[feature].median():.1f}, max={df[feature].max():.1f}")

# Check for missing values
print(f"\nMissing values in target: {df['requester_received_pizza'].isnull().sum()}")

# Text length analysis
df['request_text_length'] = df['request_text'].str.len()
df['request_title_length'] = df['request_title'].str.len()
print(f"\nText lengths - Request: avg={df['request_text_length'].mean():.0f}, Title: avg={df['request_title_length'].mean():.0f}")
```
