```
import pandas as pd
import json

# Load data
train_df = pd.read_json('/home/data/train.json')
test_df = pd.read_json('/home/data/test.json')

print("Train columns:")
print(train_df.columns.tolist())
print(f"\nTrain shape: {train_df.shape}")

print("\nTest columns:")
print(test_df.columns.tolist())
print(f"\nTest shape: {test_df.shape}")

# Check for text columns
print("\nText columns in train:")
for col in ['request_text', 'request_title', 'full_text']:
    if col in train_df.columns:
        print(f"  {col}: present")
    else:
        print(f"  {col}: NOT present")

print("\nText columns in test:")
for col in ['request_text', 'request_title', 'full_text']:
    if col in test_df.columns:
        print(f"  {col}: present")
    else:
        print(f"  {col}: NOT present")
```
