```
import pandas as pd
from catboost import CatBoostRegressor, Pool
import numpy as np

# Load data
train_df = pd.read_csv('/home/code/data/train.csv')

# Simple feature engineering
train_df['Sex'] = train_df['Sex'].map({'M': 'male', 'F': 'female'})

# Define features
feature_cols = ['Age', 'Height', 'Weight', 'Duration', 'Heart_Rate', 'Body_Temp', 'Sex']
cat_features = ['Sex']

X = train_df[feature_cols]
y = train_df['Calories']

print("Data shapes:")
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
print(f"\nX dtypes:")
print(X.dtypes)
print(f"\nFirst few rows:")
print(X.head())

# Create pool
train_pool = Pool(X, label=y, cat_features=cat_features)

# Train simple model
model = CatBoostRegressor(
    iterations=100,
    learning_rate=0.05,
    depth=4,
    verbose=False,
    random_seed=42
)

model.fit(train_pool, verbose=False)
print("\nModel trained successfully!")
```
