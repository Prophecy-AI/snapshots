```
import pandas as pd
import numpy as np
import os

data_dir = '/home/data'
full_data = pd.read_csv(os.path.join(data_dir, 'catechol_full_data_yields.csv'))
single_data = pd.read_csv(os.path.join(data_dir, 'catechol_single_solvent_yields.csv'))

print("Full Data Shape:", full_data.shape)
print("Single Data Shape:", single_data.shape)

display(full_data.head())
display(single_data.head())
```

```
lookup_files = [
    'acs_pca_descriptors_lookup.csv',
    'drfps_catechol_lookup.csv',
    'fragprints_lookup.csv',
    'spange_descriptors_lookup.csv'
]

for f in lookup_files:
    df = pd.read_csv(os.path.join(data_dir, f))
    print(f"\n{f} Shape: {df.shape}")
    display(df.head(2))
```

```
# Check target sum
full_data['yield_sum'] = full_data['SM'] + full_data['Product 2'] + full_data['Product 3']
print("Yield Sum Stats:")
print(full_data['yield_sum'].describe())

# Check correlations
print("\nTarget Correlations:")
print(full_data[['SM', 'Product 2', 'Product 3']].corr())

# Check unique solvents
print("\nUnique Solvents in Full Data (A):", full_data['SOLVENT A NAME'].nunique())
print("Unique Solvents in Full Data (B):", full_data['SOLVENT B NAME'].nunique())
print("Unique Solvents in Single Data:", single_data['SOLVENT NAME'].nunique())
```

```
all_solvents_full = set(full_data['SOLVENT A NAME']).union(set(full_data['SOLVENT B NAME']))
all_solvents_single = set(single_data['SOLVENT NAME'])

print("Solvents in Full Data:", len(all_solvents_full))
print("Solvents in Single Data:", len(all_solvents_single))
print("Missing in Full:", all_solvents_single - all_solvents_full)
print("Missing in Single:", all_solvents_full - all_solvents_single)
```
