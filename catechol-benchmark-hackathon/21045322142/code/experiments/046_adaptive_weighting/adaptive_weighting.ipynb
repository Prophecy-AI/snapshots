# Experiment 046: Adaptive Weighted Training

**Problem:** CV-LB intercept (0.0525) > Target (0.0347). Top 2 hardest solvents (Cyclohexane 35.2%, HFIP 18.6%) account for 53.8% of error.

**Hypothesis:** If we up-weight hard solvents during training, we could improve predictions for them and reduce overall error.

**Implementation:**
1. First pass: Train baseline model and compute per-solvent errors
2. Compute adaptive weights: higher weight for solvents with higher error
3. Second pass: Re-train with weighted loss
4. Test if this improves CV

```
# Standard imports
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader, WeightedRandomSampler
import lightgbm as lgb
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import Matern, WhiteKernel, ConstantKernel
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

np.random.seed(42)
torch.manual_seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')
```

```
# Data loading
DATA_PATH = '/home/data'

def load_data(name="full"):
    if name == "full":
        df = pd.read_csv(f'{DATA_PATH}/catechol_full_data_yields.csv')
        X = df[["Residence Time", "Temperature", "SOLVENT A NAME", "SOLVENT B NAME", "SolventB%"]]
    else:
        df = pd.read_csv(f'{DATA_PATH}/catechol_single_solvent_yields.csv')
        X = df[["Residence Time", "Temperature", "SOLVENT NAME"]]
    Y = df[["SM", "Product 2", "Product 3"]]
    return X, Y

# Load feature lookup tables
spange_df = pd.read_csv(f'{DATA_PATH}/spange_descriptors_lookup.csv', index_col=0)
drfp_df = pd.read_csv(f'{DATA_PATH}/drfps_catechol_lookup.csv', index_col=0)

SPANGE_COLS = [c for c in spange_df.columns if c != 'solvent smiles']
DRFP_COLS = [c for c in drfp_df.columns if str(c).isdigit() or isinstance(c, int)]

print(f'Spange: {len(SPANGE_COLS)} features')
print(f'DRFP: {len(DRFP_COLS)} features')

# Load data
X_single, Y_single = load_data('single_solvent')
X_full, Y_full = load_data('full')

print(f'Single solvent: {len(X_single)} samples')
print(f'Full data: {len(X_full)} samples')
```

```
# Feature extraction (baseline from exp_030)
def get_features(X, data_type='single'):
    features_list = []
    
    for idx, row in X.iterrows():
        time_m = row['Residence Time']
        temp_c = row['Temperature']
        temp_k = temp_c + 273.15
        
        kinetics = np.array([
            time_m, temp_c, 1.0 / temp_k,
            np.log(time_m + 1), time_m / temp_k
        ], dtype=np.float32)
        
        if data_type == 'single':
            solvent = row['SOLVENT NAME']
            spange = spange_df.loc[solvent, SPANGE_COLS].values.astype(np.float32) if solvent in spange_df.index else np.zeros(len(SPANGE_COLS), dtype=np.float32)
            drfp = drfp_df.loc[solvent, DRFP_COLS].values.astype(np.float32) if solvent in drfp_df.index else np.zeros(len(DRFP_COLS), dtype=np.float32)
        else:
            solvent_a = row['SOLVENT A NAME']
            solvent_b = row['SOLVENT B NAME']
            pct_b = row['SolventB%'] / 100.0
            pct_a = 1 - pct_b
            
            sp_a = spange_df.loc[solvent_a, SPANGE_COLS].values.astype(np.float32) if solvent_a in spange_df.index else np.zeros(len(SPANGE_COLS), dtype=np.float32)
            sp_b = spange_df.loc[solvent_b, SPANGE_COLS].values.astype(np.float32) if solvent_b in spange_df.index else np.zeros(len(SPANGE_COLS), dtype=np.float32)
            spange = pct_a * sp_a + pct_b * sp_b
            
            dr_a = drfp_df.loc[solvent_a, DRFP_COLS].values.astype(np.float32) if solvent_a in drfp_df.index else np.zeros(len(DRFP_COLS), dtype=np.float32)
            dr_b = drfp_df.loc[solvent_b, DRFP_COLS].values.astype(np.float32) if solvent_b in drfp_df.index else np.zeros(len(DRFP_COLS), dtype=np.float32)
            drfp = pct_a * dr_a + pct_b * dr_b
        
        features = np.concatenate([kinetics, spange, drfp])
        features_list.append(features)
    
    return np.array(features_list, dtype=np.float32)

print('Feature extraction defined')
```

```
# MLP Model with weighted loss
class MLPModel(nn.Module):
    def __init__(self, input_dim, hidden_dims=[32, 16]):
        super().__init__()
        layers = []
        prev_dim = input_dim
        for h_dim in hidden_dims:
            layers.extend([
                nn.Linear(prev_dim, h_dim),
                nn.BatchNorm1d(h_dim),
                nn.ReLU(),
                nn.Dropout(0.3)
            ])
            prev_dim = h_dim
        layers.append(nn.Linear(prev_dim, 3))
        self.network = nn.Sequential(*layers)
    
    def forward(self, x):
        return self.network(x)

print('MLPModel defined')
```

```
# Adaptive Weighted Model
class AdaptiveWeightedModel:
    """Model that uses adaptive weighting to focus on hard solvents.
    
    The idea is to up-weight samples from solvents that are hard to predict,
    so the model focuses more on getting those right.
    """
    
    def __init__(self, data='single', gp_weight=0.15, mlp_weight=0.55, lgbm_weight=0.3,
                 weighting_power=0.5):  # sqrt weighting by default
        self.data_type = data
        self.gp_weight = gp_weight
        self.mlp_weight = mlp_weight
        self.lgbm_weight = lgbm_weight
        self.weighting_power = weighting_power
        
        self.scaler = None
        self.gp_models = []
        self.mlp_models = []
        self.lgbm_models = []
        self.sample_weights = None
    
    def _compute_sample_weights(self, X_train, y_train, X_feat_scaled):
        """Compute sample weights based on prediction difficulty."""
        # First, train a quick model to estimate errors
        quick_lgbm = lgb.LGBMRegressor(n_estimators=50, learning_rate=0.1, max_depth=3, verbose=-1)
        quick_lgbm.fit(X_feat_scaled, y_train.values[:, 0])  # Just use first target
        
        # Predict on training data
        preds = quick_lgbm.predict(X_feat_scaled)
        errors = np.abs(preds - y_train.values[:, 0])
        
        # Compute per-solvent average error
        if self.data_type == 'single':
            solvents = X_train['SOLVENT NAME'].values
        else:
            solvents = X_train['SOLVENT A NAME'].values
        
        solvent_errors = {}
        for solvent in np.unique(solvents):
            mask = solvents == solvent
            solvent_errors[solvent] = errors[mask].mean()
        
        # Compute weights: higher weight for higher error solvents
        mean_error = np.mean(list(solvent_errors.values()))
        weights = np.array([solvent_errors[s] / mean_error for s in solvents])
        
        # Apply power transformation (sqrt by default)
        weights = np.power(weights, self.weighting_power)
        
        # Normalize to mean 1
        weights = weights / weights.mean()
        
        return weights
    
    def train_model(self, X_train, y_train, epochs=200):
        X_feat = get_features(X_train, self.data_type)
        y_np = y_train.values.astype(np.float32)
        
        # Scale features
        self.scaler = StandardScaler()
        X_scaled = self.scaler.fit_transform(X_feat)
        
        # Compute sample weights
        self.sample_weights = self._compute_sample_weights(X_train, y_train, X_scaled)
        
        # Train GP models (one per target) - GP doesn't support sample weights easily
        self.gp_models = []
        for i in range(3):
            kernel = ConstantKernel(1.0) * Matern(length_scale=1.0, nu=2.5) + WhiteKernel(noise_level=0.1)
            gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=2, random_state=42)
            gp.fit(X_scaled[:, :18], y_np[:, i])
            self.gp_models.append(gp)
        
        # Train MLP models with weighted loss
        self.mlp_models = []
        for _ in range(3):
            model = MLPModel(X_scaled.shape[1], hidden_dims=[32, 16]).to(device)
            optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-4)
            scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)
            
            X_tensor = torch.tensor(X_scaled).to(device)
            y_tensor = torch.tensor(y_np).to(device)
            weights_tensor = torch.tensor(self.sample_weights, dtype=torch.float32).to(device)
            
            dataset = TensorDataset(X_tensor, y_tensor, weights_tensor.unsqueeze(1))
            loader = DataLoader(dataset, batch_size=32, shuffle=True)
            
            model.train()
            for epoch in range(epochs):
                for X_batch, y_batch, w_batch in loader:
                    optimizer.zero_grad()
                    pred = model(X_batch)
                    # Weighted MSE loss
                    target_weights = torch.tensor([1.0, 1.0, 2.0]).to(device)
                    loss = (w_batch * target_weights * (pred - y_batch)**2).mean()
                    loss.backward()
                    optimizer.step()
                scheduler.step()
            
            model.eval()
            self.mlp_models.append(model)
        
        # Train LGBM models with sample weights
        self.lgbm_models = []
        for i in range(3):
            lgbm_model = lgb.LGBMRegressor(
                n_estimators=100, learning_rate=0.05, max_depth=5,
                num_leaves=31, random_state=42, verbose=-1
            )
            lgbm_model.fit(X_scaled, y_np[:, i], sample_weight=self.sample_weights)
            self.lgbm_models.append(lgbm_model)
        
        return self
    
    def predict(self, X_test):
        X_feat = get_features(X_test, self.data_type)
        X_scaled = self.scaler.transform(X_feat)
        
        # GP predictions
        gp_preds = np.zeros((len(X_test), 3))
        for i, gp in enumerate(self.gp_models):
            gp_preds[:, i] = gp.predict(X_scaled[:, :18])
        
        # MLP predictions
        mlp_preds = []
        for model in self.mlp_models:
            X_tensor = torch.tensor(X_scaled).to(device)
            with torch.no_grad():
                pred = model(X_tensor).cpu().numpy()
            mlp_preds.append(pred)
        mlp_preds = np.mean(mlp_preds, axis=0)
        
        # LGBM predictions
        lgbm_preds = np.zeros((len(X_test), 3))
        for i, lgbm_model in enumerate(self.lgbm_models):
            lgbm_preds[:, i] = lgbm_model.predict(X_scaled)
        
        # Ensemble
        ensemble_preds = self.gp_weight * gp_preds + self.mlp_weight * mlp_preds + self.lgbm_weight * lgbm_preds
        ensemble_preds = np.clip(ensemble_preds, 0, 1)
        
        return torch.tensor(ensemble_preds, dtype=torch.float32)

print('AdaptiveWeightedModel defined')
```

```
# Test adaptive weighting on single solvent data
print("Testing adaptive weighted training on single solvent data...")
print()

all_solvents = sorted(X_single["SOLVENT NAME"].unique())

# Test different weighting powers
power_results = {}
for power in [0.0, 0.25, 0.5, 0.75, 1.0]:  # 0.0 = no weighting (baseline)
    fold_mses = []
    
    for test_solvent in all_solvents:
        mask = X_single["SOLVENT NAME"] != test_solvent
        
        model = AdaptiveWeightedModel(data='single', weighting_power=power)
        model.train_model(X_single[mask], Y_single[mask], epochs=150)
        preds = model.predict(X_single[~mask])
        
        actuals = Y_single[~mask].values
        mse = np.mean((actuals - preds.numpy())**2)
        fold_mses.append(mse)
    
    mean_mse = np.mean(fold_mses)
    std_mse = np.std(fold_mses)
    power_results[power] = (mean_mse, std_mse)
    print(f"Power={power:.2f}: CV MSE = {mean_mse:.6f} +/- {std_mse:.6f}")

print()
print("Baseline (exp_030): CV = 0.008298")
```

```
# Find best power
best_power = min(power_results.keys(), key=lambda p: power_results[p][0])
best_mse = power_results[best_power][0]

print(f"Best power: {best_power}")
print(f"Best CV MSE: {best_mse:.6f}")
print(f"Baseline (exp_030): CV = 0.008298")
print(f"Improvement: {(0.008298 - best_mse) / 0.008298 * 100:.1f}%")
```
