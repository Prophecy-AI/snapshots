```
import pandas as pd
import numpy as np

# Load data
full_data = pd.read_csv('/home/data/catechol_full_data_yields.csv')
single_data = pd.read_csv('/home/data/catechol_single_solvent_yields.csv')

print('Full data shape:', full_data.shape)
print('Single solvent data shape:', single_data.shape)
print('\nFull data columns:', full_data.columns.tolist())
print('\nSingle data columns:', single_data.columns.tolist())
```

```
# Check target distributions and unique solvents
print('Target columns (SM, Product 2, Product 3):')
print(single_data[['SM', 'Product 2', 'Product 3']].describe())

print('\nUnique solvents in single data:', single_data['SOLVENT NAME'].nunique())
print('Solvents:', single_data['SOLVENT NAME'].unique())

print('\nUnique solvent pairs in full data:')
print('Solvent A:', full_data['SOLVENT A NAME'].nunique())
print('Solvent B:', full_data['SOLVENT B NAME'].nunique())

# Check SolventB% range
print('\nSolventB% range:', full_data['SolventB%'].min(), '-', full_data['SolventB%'].max())
```

```
# Check lookup tables
spange = pd.read_csv('/home/data/spange_descriptors_lookup.csv', index_col=0)
acs_pca = pd.read_csv('/home/data/acs_pca_descriptors_lookup.csv', index_col=0)
drfps = pd.read_csv('/home/data/drfps_catechol_lookup.csv', index_col=0)
fragprints = pd.read_csv('/home/data/fragprints_lookup.csv', index_col=0)

print('Spange descriptors shape:', spange.shape)
print('ACS PCA descriptors shape:', acs_pca.shape)
print('DRFP shape:', drfps.shape)
print('Fragprints shape:', fragprints.shape)

print('\nSpange columns:', spange.columns.tolist())
print('\nACS PCA columns:', acs_pca.columns.tolist())
```

```
# Check Temperature and Residence Time ranges
print('Temperature range:', single_data['Temperature'].min(), '-', single_data['Temperature'].max())
print('Residence Time range:', single_data['Residence Time'].min(), '-', single_data['Residence Time'].max())

# Check CV structure - how many samples per solvent
print('\nSamples per solvent (single data):')
print(single_data.groupby('SOLVENT NAME').size().describe())

# Check number of ramps in full data
print('\nNumber of unique ramps (solvent pairs) in full data:')
ramps = full_data[['SOLVENT A NAME', 'SOLVENT B NAME']].drop_duplicates()
print(len(ramps))
```
