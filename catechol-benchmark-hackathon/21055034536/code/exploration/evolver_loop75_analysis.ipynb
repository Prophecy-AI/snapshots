# Loop 75 Analysis  Submission failure: "Evaluation metric raised an unexpected error"

We need to restore *scoring* behavior before any more submissions.

The most likely root cause is **submission file path mismatch**.

- Official template writes: `submission.to_csv("submission.csv", index=True)`
- Our notebooks write: `submission.to_csv("/home/submission/submission.csv", index=True)`

If the evaluator looks for a hardcoded `submission.csv` in the notebook working directory, it will crash (metric cannot load preds) and Kaggle surfaces this as "evaluation metric raised an unexpected error".

This loop verifies the path difference between template and exp_075, and sets the fix policy: **always write exactly `submission.csv`**.


```
import nbformat
from pathlib import Path

def last_cell_sources(nb_path, n=3):
    nb = nbformat.read(nb_path, as_version=4)
    return [nb.cells[i].source for i in range(len(nb.cells)-n, len(nb.cells))]

exp075_nb = Path('/home/code/experiments/075_AD_shrinkage_v3_flip_invariant/ad_shrinkage_v3_flip_invariant.ipynb')
template_nb = Path('research/kernels/josepablofolch_catechol-benchmark-hackathon-template/catechol-benchmark-hackathon-template.ipynb')

exp075_last3 = last_cell_sources(exp075_nb, 3)
template_last3 = last_cell_sources(template_nb, 3)

print('--- exp_075 final cell ---')
print(exp075_last3[-1])
print('\n--- template final cell ---')
print(template_last3[-1])

print('\nPath difference check:')
print('exp_075 uses /home/submission?', '/home/submission' in exp075_last3[-1])
print('template uses to_csv("submission.csv")?', 'to_csv("submission.csv"' in template_last3[-1])
```
