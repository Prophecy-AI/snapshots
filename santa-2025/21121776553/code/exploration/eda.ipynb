```
import pandas as pd
import numpy as np

# Load sample submission
df = pd.read_csv('/home/data/sample_submission.csv')
print(f"Shape: {df.shape}")
print(f"\nFirst 10 rows:")
print(df.head(10))
print(f"\nLast 10 rows:")
print(df.tail(10))
```

```
# Analyze score contribution by N\ndf['n'] = df['id'].apply(lambda x: int(x.split('_')[0]))\ndf['x_val'] = df['x'].str[1:].astype(float)\ndf['y_val'] = df['y'].str[1:].astype(float)\n\n# Calculate bounding box and score for each N\nscores = []\nfor n in range(1, 201):\n    group = df[df['n'] == n]\n    x_min, x_max = group['x_val'].min(), group['x_val'].max()\n    y_min, y_max = group['y_val'].min(), group['y_val'].max()\n    # Note: This is approximate - actual score uses polygon bounds, not center points\n    side = max(x_max - x_min, y_max - y_min)\n    score = side**2 / n\n    scores.append({'n': n, 'side': side, 'score': score})\n\nscores_df = pd.DataFrame(scores)\nprint(f\"Total score (approximate, based on centers): {scores_df['score'].sum():.6f}\")\nprint(f\"\\nTop 10 highest score contributions:\")\nprint(scores_df.nlargest(10, 'score'))\nprint(f\"\\nScore contribution by N range:\")\nfor start in [1, 51, 101, 151]:\n    end = start + 49\n    subset = scores_df[(scores_df['n'] >= start) & (scores_df['n'] <= end)]\n    print(f\"  N={start}-{end}: {subset['score'].sum():.4f} ({100*subset['score'].sum()/scores_df['score'].sum():.1f}%)\")
```
