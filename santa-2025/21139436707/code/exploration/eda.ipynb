# Quick Data Exploration - Christmas Tree Packing

```
import pandas as pd
import numpy as np

# Load sample submission to understand data format
df = pd.read_csv('/home/data/sample_submission.csv')
print(f"Shape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nFirst 10 rows:")
df.head(10)
```

```
# Understand the structure: 200 configurations with 1 to 200 trees each
# Total rows = 1 + 2 + 3 + ... + 200 = 200*201/2 = 20100

# Check the distribution of configurations
df['n_trees'] = df['id'].apply(lambda x: int(x.split('_')[0]))
df['tree_idx'] = df['id'].apply(lambda x: int(x.split('_')[1]))

print(f"Number of configurations: {df['n_trees'].nunique()}")
print(f"Min trees: {df['n_trees'].min()}, Max trees: {df['n_trees'].max()}")
print(f"\nTotal expected rows: {200*201//2}")
print(f"Actual rows: {len(df)}")
```
