```
import json

with open('/home/code/research/kernels/smartmanoj_santa-claude/santa-claude.ipynb', 'r') as f:
    nb = json.load(f)

for cell in nb['cells']:
    if cell['cell_type'] == 'code':
        source = ''.join(cell['source'])
        if '%%writefile' in source:
            print(f"--- Found writefile in cell ---")
            print(source[:100]) # Print first 100 chars to identify
            filename = source.split()[1]
            print(f"File: {filename}")
            
            # Save to file
            with open(filename, 'w') as out:
                # Remove the %%writefile line
                content = '\n'.join(source.split('\n')[1:])
                out.write(content)
            print(f"Saved {filename}")
```
