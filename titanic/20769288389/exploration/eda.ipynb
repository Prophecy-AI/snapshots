```
import pandas as pd
import numpy as np

# Load data
train = pd.read_csv('/home/data/train.csv')
test = pd.read_csv('/home/data/test.csv')

print(f"Train shape: {train.shape}")
print(f"Test shape: {test.shape}")
print("\nTrain columns:")
print(train.dtypes)
```

```
# Check missing values
print("Missing values in train:")
print(train.isnull().sum())
print("\nMissing values in test:")
print(test.isnull().sum())

print("\n\nTarget distribution:")
print(train['Survived'].value_counts())
print(f"\nSurvival rate: {train['Survived'].mean():.2%}")
```

```
# Key feature analysis
print("Survival by Sex:")
print(train.groupby('Sex')['Survived'].agg(['mean', 'count']))

print("\nSurvival by Pclass:")
print(train.groupby('Pclass')['Survived'].agg(['mean', 'count']))

print("\nSurvival by Embarked:")
print(train.groupby('Embarked')['Survived'].agg(['mean', 'count']))
```
