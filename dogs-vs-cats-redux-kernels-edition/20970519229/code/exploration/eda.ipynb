```
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

# Check data structure
train_dir = '/home/data/train/'
test_dir = '/home/data/test/'

print("Training images:", len(os.listdir(train_dir)))
print("Test images:", len(os.listdir(test_dir)))

# Check a few sample images
sample_images = os.listdir(train_dir)[:10]
print("\nSample training images:")
for img in sample_images:
    print(img)
```

```
# Analyze class distribution
train_images = os.listdir(train_dir)
dog_images = [img for img in train_images if img.startswith('dog')]
cat_images = [img for img in train_images if img.startswith('cat')]

print(f"Total training images: {len(train_images)}")
print(f"Dog images: {len(dog_images)}")
print(f"Cat images: {len(cat_images)}")
print(f"Class balance: {len(dog_images)/len(train_images):.2%} dogs, {len(cat_images)/len(train_images):.2%} cats")

# Check image sizes and formats
sample_img_path = os.path.join(train_dir, train_images[0])
img = Image.open(sample_img_path)
print(f"\nSample image format: {img.format}")
print(f"Sample image size: {img.size}")
print(f"Sample image mode: {img.mode}")
```
