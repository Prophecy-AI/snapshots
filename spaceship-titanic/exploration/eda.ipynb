```
import pandas as pd
import numpy as np

# Load data
train = pd.read_csv('/home/data/train.csv')
test = pd.read_csv('/home/data/test.csv')

print(f'Train shape: {train.shape}')
print(f'Test shape: {test.shape}')
print('\nColumns:', train.columns.tolist())
print('\nData types:')
print(train.dtypes)
```

```
# Check target distribution and missing values
print('Target distribution:')
print(train['Transported'].value_counts(normalize=True))
print('\nMissing values:')
print(train.isnull().sum())
print('\nSample data:')
train.head()
```

```
# Key feature analysis
print('CryoSleep vs Transported:')
print(pd.crosstab(train['CryoSleep'], train['Transported'], normalize='index'))

print('\nHomePlanet vs Transported:')
print(pd.crosstab(train['HomePlanet'], train['Transported'], normalize='index'))

print('\nSpending features statistics:')
spending_cols = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
print(train[spending_cols].describe())
```
